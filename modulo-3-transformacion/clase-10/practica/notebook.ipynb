### Ejercicio 5 - Obtén para cada tipo de contenido (Movie y TV Show):

- cantidad de títulos
- año promedio de lanzamiento
- año más reciente
- año más antiguo

> Utiliza **agg()**.

```python
df.groupby("release_year").agg(
    Cantidad_Tipo_año   = ("type",    "sum"),
    Promedio_tipo= ("type",    "mean"),
    reciente_max  = ("release_year","max"),
    reciente_min  = ("release_year","min")
)
```
